```python
import pandas as pd
import sklearn.datasets as ds
import numpy as np
import sklearn.model_selection as ms
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
```

Generiamo il dataset

```python
X, y = make_moons(n_samples=500, noise=0.05, random_state=40)
```

```python
fig = plt.figure(figsize = [12, 5])
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.scatter(X[:, 0], X[:, 1], c=y, cmap='rainbow')
ax2.scatter(X[:, 0], X[:, 1], cmap='rainbow')
ax1.set_title('Dati con etichetta fittizia')
ax2.set_title('Dati non etichettati')
plt.show()
```

Addestriamo il modello con il K-means

```python
kmeans = KMeans(init="random", n_clusters=2, n_init=10, random_state=30)
kmeans.fit(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow')
```

Addestriamo il modello con il DBSCAN

```python
epsilon = 0.2
min_samples = 5

db = DBSCAN(eps=epsilon, min_samples=min_samples)
db.fit(X)
labels = db.labels_
```

```python
y_pred = db.fit_predict(X)
plt.scatter(X[:,0], X[:,1],c=y_pred, cmap='Paired')
plt.title("DBSCAN")
```
